```python
# prompt: Track items in stock:
# Fields: Item ID, Item Name, Category, Quantity, Supplier Name, Reorder Level.
# Add or update inventory items (e.g., quantity or reorder level).

import pandas as pd

def update_inventory(df, item_id, quantity_change=0, reorder_level_change=0):
    """
    Updates the quantity and/or reorder level of an item in the inventory.

    Args:
        df: The pandas DataFrame representing the inventory.
        item_id: The ID of the item to update.
        quantity_change: The change in quantity (can be positive or negative).
        reorder_level_change: The change in reorder level.

    Returns:
      Updated DataFrame or None if item_id not found
    """
    try:
        # Attempt to locate the item by its ID.
        item_index = df.index[df['Item ID'] == item_id].tolist()[0]

        # Update quantity and reorder level.
        df.loc[item_index, 'Quantity'] += quantity_change
        df.loc[item_index, 'Reorder Level'] += reorder_level_change

        return df

    except IndexError:
        print(f"Item with ID '{item_id}' not found in inventory.")
        return None





```

```python
# prompt: Track Orders
# Record new orders:
# Fields: Order ID, Item ID, Quantity Ordered, Order Date, Delivery Status
# (Pending/Delivered).
# Update delivery status for existing orders.

import pandas as pd

def record_new_order(orders_df, order_id, item_id, quantity, order_date, delivery_status):
    """Records a new order in the DataFrame."""
    new_order = pd.DataFrame({
        'Order ID': [order_id],
        'Item ID': [item_id],
        'Quantity Ordered': [quantity],
        'Order Date': [order_date],
        'Delivery Status': [delivery_status]
    })
    return pd.concat([orders_df, new_order], ignore_index=True)



```

```python
def update_delivery_status(orders_df, order_id, new_status):
    """Updates the delivery status of an existing order."""
    try:
        order_index = orders_df.index[orders_df['Order ID'] == order_id].tolist()[0]
        orders_df.loc[order_index, 'Delivery Status'] = new_status
        return orders_df
    except IndexError:
        print(f"Order with ID '{order_id}' not found.")
        return orders_df
```

```python
# Example usage:
# Initialize an empty orders DataFrame (or load from a file)
orders_data = {'Order ID': [], 'Item ID': [], 'Quantity Ordered': [], 'Order Date': [], 'Delivery Status': []}
orders_df = pd.DataFrame(orders_data)

# Record new orders
orders_df = record_new_order(orders_df, 'ORDER001', 'ITEM001', 5, '2024-03-08', 'Pending')
orders_df = record_new_order(orders_df, 'ORDER002', 'ITEM002', 2, '2024-03-09', 'Pending')

# Update delivery status
orders_df = update_delivery_status(orders_df, 'ORDER001', 'Delivered')

orders_df

# Save to a CSV file (optional)
# orders_df.to_csv("orders.csv", index=False)
```

```python
# prompt:Generate Reports
# Display a list of low-stock items (items below the reorder level).
# View pending orders.
# Summarize total stock by category

def display_low_stock(df):
    """Displays a list of low-stock items."""
    low_stock_items = df[df['Quantity'] < df['Reorder Level']]
    return low_stock_items



```

```python
def view_pending_orders(orders_df):
    """Displays a list of pending orders."""
    pending_orders = orders_df[orders_df['Delivery Status'] == 'Pending']
    return pending_orders
```

```python
def summarize_stock_by_category(df):
    """Summarizes total stock by category."""
    stock_summary = df.groupby('Category')['Quantity'].sum()
    return stock_summary

# Example Usage
low_stock = display_low_stock(df)
print("Low Stock Items:\n", low_stock)
pending = view_pending_orders(orders_df)
print("\nPending Orders:\n", pending)

stock_summary = summarize_stock_by_category(df)
print("\nStock Summary by Category:\n", stock_summary)
```
